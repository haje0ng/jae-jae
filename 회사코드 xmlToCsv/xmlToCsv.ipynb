```
path = "drive/My Drive/Colab Notebooks/ios/xmlToCsv/data/"
```

```
from xml.etree.ElementTree import parse
import pandas as pd
```

```
tree = parse(path + "CORPCODE.xml")
root = tree.getroot()
df = []
for element in root.findall("list"):
    corp_code = element.findtext("corp_code")
    corp_name = element.findtext("corp_name")
    modify_date = element.findtext("modify_date")
    df.append({"corp_code":corp_code, "corp_name":corp_name, "modify_date":modify_date})
```

```
df = pd.DataFrame(df)
```

```
print(df.shape)
df
```

```
df.to_csv(path + "corpcode.csv")
```
